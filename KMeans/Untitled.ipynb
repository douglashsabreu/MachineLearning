```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerando dados através de três distribuição gaussianas
dados1 = np.random.uniform(0, 0.2, (100,2))
dados2 = np.random.uniform(1, 0.6, (100,2))
dados3 = np.random.uniform(0.72, 0.2, (100,2))

dados = np.concatenate((dados1, dados2, dados3))
```

```python
plt.title('Dados originais')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(dados1[:,0], dados1[:,1])
plt.scatter(dados2[:,0], dados2[:,1])
plt.scatter(dados3[:,0], dados3[:,1])
```

```python
# Gerando clusters
cluster_number = 3
# Gera K centróides de dimensão 1x2
clusters_centroids = np.random.random((cluster_number,2))
plt.title('Dados sem rotulados e centróides posicionadas')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(dados[:,0], dados[:,1], alpha=0.5, color='blue')
plt.scatter(clusters_centroids[:,0], clusters_centroids[:,1], marker='o', s=100, color='red')
```

```python
def computar_distancia_euclidiana(x,y):
    return np.sqrt((x[0] - y[0]) ** 2 + (x[1] - y[1]) ** 2)

for i in range(10):
    dados_cluster1, dados_cluster2, dados_cluster3  = [], [], []

    for sample in dados:
        # Computa a distância do exemplo até o centróide do cluster 1
        dist_1 = computar_distancia_euclidiana(sample, clusters_centroids[0])
        dist_2 = computar_distancia_euclidiana(sample, clusters_centroids[1])
        dist_3 = computar_distancia_euclidiana(sample, clusters_centroids[2])


        if dist_1 <= dist_2 and dist_1 <= dist_3:
            dados_cluster1.append(sample)
        elif dist_2 <= dist_1 and dist_2 <= dist_3:
            dados_cluster2.append(sample)
        else:
            dados_cluster3.append(sample)

    plt.figure(figsize=(15,10))
    plt.subplot(2,2,1)
    plt.title('Clusterização dos dados. Iteração '+ str(i + 1))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.scatter(np.array(dados_cluster1)[:,0], np.array(dados_cluster1)[:,1], alpha=0.5)
    plt.scatter(np.array(dados_cluster2)[:,0], np.array(dados_cluster2)[:,1], alpha=0.5)
    plt.scatter(np.array(dados_cluster3)[:,0], np.array(dados_cluster3)[:,1], alpha=0.5)
    plt.scatter(clusters_centroids[:,0], clusters_centroids[:,1], marker='o', s=100, color='red')

    clusters_centroids[0] = np.mean(dados_cluster1, axis=0)
    clusters_centroids[1] = np.mean(dados_cluster2, axis=0)
    clusters_centroids[2] = np.mean(dados_cluster3, axis=0)

    plt.subplot(2,2,2)
    plt.title('Realocação dos centróides. Iteração '+ str(i + 1))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.scatter(np.array(dados_cluster1)[:,0], np.array(dados_cluster1)[:,1], alpha=0.5)
    plt.scatter(np.array(dados_cluster2)[:,0], np.array(dados_cluster2)[:,1], alpha=0.5)
    plt.scatter(np.array(dados_cluster3)[:,0], np.array(dados_cluster3)[:,1], alpha=0.5)
    plt.scatter(clusters_centroids[:,0], clusters_centroids[:,1], marker='o', s=100, color='red')
    plt.plot()
```
